```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
```

```python
language = ['en', 'fr', 'de', 'es']
domain = ['animals', 'plants', 'arts']
for d in domain:
    sections = pd.DataFrame(columns=language)
    for l in language:
        temp = pd.read_csv(d+'_'+l+'_top20_sections.csv')
        sections[l] = temp['section']
    sections.to_csv(d+'_top20_sections.csv')
```

```python
# entry = pd.read_csv('animals_en_info.csv', index_col=0)
for d in domain:
    sections = pd.DataFrame(columns=language)
    for l in language:
        temp = pd.read_csv(d+'_'+l+'_info.csv')
        temp.set_index('data', inplace=True)
        sections[l] = temp['value']
    sections = pd.DataFrame(sections.values.T, index=sections.columns, columns=sections.index)
    sections['domain'] = d
    sections.to_csv(d+'_info.csv')
```

```python
animal_info = pd.read_csv('animals_info.csv', index_col=0)
plant_info = pd.read_csv('plants_info.csv', index_col=0)
art_info = pd.read_csv('arts_info.csv', index_col=0)
art_info.loc['fr'] = 0
art_info.domain['fr'] = 'arts'
art_info
```

```python
info = pd.concat([animal_info, plant_info, art_info])
info.reset_index(inplace=True)
info
```

```python
sns.catplot(data=info, x='domain', y='page_num', hue='index', kind='bar')
plt.title('Total Page Number')
plt.show()
```

```python
sns.catplot(data=info, x='domain', y='avg_section_num', hue='index', kind='bar')
plt.title('Average Section Number')
plt.show()
```

```python
sns.catplot(data=info, x='domain', y='avg_reference_num', hue='index', kind='bar')
plt.title('Average Reference Number')
plt.show()
```

```python
sns.catplot(data=info, x='domain', y='avg_reference_5prop_num', hue='index', kind='bar')
plt.title('Proportion of Reference Pages with > 5 Paragraphs')
plt.show()
```

```python
sns.catplot(data=info, x='domain', y='scraping_time', hue='index', kind='bar')
plt.title('Average Scraping Time')
plt.show()
```

```python
sns.catplot(data=info, x='domain', y='problem_page_num', hue='index', kind='bar')
plt.title('Number of Problem Pages')
plt.show()
```
